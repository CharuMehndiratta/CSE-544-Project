```python
ru_df = pd.DataFrame(review_df[review_df['business_id'] == 's_466owgNXqmQRuSluHHEQ'])
len(ru_df)
```

```python
reviews_per_year = {}

for year in range(2007, 2020):
    avg = 0.0
    count = 0
    for index, r in ru_df.iterrows():
        if (pd.to_datetime(r['date']).year == year):
            avg = avg + r['stars']
            count = count + 1
        if (count > 0):
            reviews_per_year[year] = float(avg/count)
        else :
            reviews_per_year[year] = 0.0
print(reviews_per_year)
```

```python
y_label = reviews_per_year.values()
x_label = reviews_per_year.keys()
    
plt.scatter(
    x_label,
    y_label,
    color='b')
plt.show()
```
